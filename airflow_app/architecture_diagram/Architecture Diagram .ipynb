```python
from diagrams import Diagram
from diagrams.aws.storage import S3
from diagrams.onprem.client import Users
from diagrams.custom import Custom
from diagrams.onprem.workflow import Airflow as AirflowIcon

# Paths to custom icons
streamlit_icon = "./Streamlit.png"
fastapi_icon = "./fastAPI.png"
airflow_icon = "./airflow.png"
snowflake_icon = "./snowflake.png"
dbt_icon = "./dbt.png"  # Replace with the actual path to your DBT icon

with Diagram("End-to-End Data Pipeline with DBT and Snowflake", show=False) as diag:
    end_user = Users("End User\n(Streamlit App)")
    streamlit_app = Custom("Streamlit App", streamlit_icon)
    fastapi_service_1 = Custom("FastAPI Service #1\n(File Upload)", fastapi_icon)
    s3_bucket = S3("S3 Bucket\n(PDF Storage)")
    snowflake_storage = Custom("Snowflake\n(Data & Metadata Storage)", snowflake_icon)
    airflow_orchestrator = Custom("Airflow Orchestration", airflow_icon)
    dbt_transformation = Custom("DBT Transformation", dbt_icon)
    fastapi_service_2 = Custom("FastAPI Service #2\n(Query Interface)", fastapi_icon)

    end_user >> streamlit_app >> fastapi_service_1 >> s3_bucket >> snowflake_storage
    snowflake_storage >> airflow_orchestrator >> dbt_transformation
    end_user >> fastapi_service_2 >> snowflake_storage

diag

```
